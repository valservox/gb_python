```python
import pandas as pd
import numpy as np
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'who_am_i':lst})
data
```

```python
data['human'] = np.where(data['who_am_i'] == 'human', 1, 0)
data['robot'] = np.where(data['who_am_i'] == 'robot', 1, 0)
data.drop(['who_am_i'],axis=1)
```
